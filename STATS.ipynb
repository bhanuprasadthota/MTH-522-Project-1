```python
import pandas as pd
import statsmodels.api as sm
```

```python
X=sm.add_constant(do["% OBESE"].values.reshape(-1, 1))
y=(do["% DIABETIC"])
m1= sm.OLS(y,X).fit()
print(len(X))
print(len(y))
print(m1.summary())
intercept = m1.params['const']
print(f"Intercept: {intercept}")

```

```python
y_pred = m1.predict(X)
```

```python
plt.scatter(do["% OBESE"],y,label='OBSE TO DIABETES', color='blue')
plt.plot(do["% OBESE"], y_pred, label='OLS Regression Line', color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
X1=sm.add_constant(oi["% INACTIVE"].values.reshape(-1, 1))
y1=oi["% OBESE"]
m2= sm.OLS(y1,X1).fit()
print(m2.summary())
intercept = m2.params['const']
print(f"Intercept: {intercept}")

print(len(X1))
print(len(y1))
y1_pred = m2.predict(X1)

plt.scatter(oi["% INACTIVE"],y1,label='INACTIVE TO OBESITY', color='blue')
plt.plot(oi["% INACTIVE"], y1_pred, label='OLS Regression Line', color='red')


plt.xlabel('X1')
plt.ylabel('y1')
plt.legend()
plt.show()
```

```python
X3=sm.add_constant(di["% INACTIVE"])
y3=di["% DIABETIC"]
m3= sm.OLS(y3, X3).fit()
print(m3.summary())
intercept = m3.params['const']
print(f"Intercept: {intercept}")
```

```python
print(len(X3))
print(len(y3))
y3_pred = m3.predict(X3)

plt.scatter(di["% INACTIVE"],y3,label='INACTIVE TO Diabetics', color='blue')
plt.plot(di["% INACTIVE"], y3_pred, label='OLS Regression Line', color='red')


plt.xlabel('X3')
plt.ylabel('y3')
plt.legend()
plt.show()
```

```python
X31=di["% INACTIVE"]
y31=sm.add_constant(di["% DIABETIC"])
m31= sm.OLS(X31, y31).fit()
print(m31.summary())
intercept = m31.params['const']
print(f"Intercept: {intercept}")

print(len(X31))
print(len(y31))
X31_pred = m31.predict(y31)

plt.scatter(di["% DIABETIC"],X31,label='INACTIVE TO Diabetics', color='blue')
plt.plot(di["% DIABETIC"], X31_pred, label='OLS Regression Line', color='red')


plt.xlabel('X31')
plt.ylabel('y31')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
ols_model = LinearRegression()


X1 = oi["% INACTIVE"].values.reshape(-1, 1)
y1 = oi["% OBESE"]
# Fit the model to your data
ols_model.fit(X1, y1)

# Predict Y values using the model
Y_pred4 = ols_model.predict(X1)

# Create a scatter plot of the data
plt.scatter(oi["% INACTIVE"], y1, label='Data', color='blue')

# Plot the regression line
plt.plot(oi["% INACTIVE"], Y_pred4, label='Regression Line', color='red')

# Add labels and a legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Show the plot
plt.show()
```
